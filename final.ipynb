```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('universal_tagset')
from textblob import TextBlob
from collections import Counter
from nltk.corpus import wordnet as wn
```

```python
file_path1 = os.path.join("data","jobtitle", "jobss.csv")
job = pd.read_csv(file_path1)
```

```python
job.drop(columns=['Job Experience Required', 'Unnamed: 1','Location','Longitude','Latitude','sal'],inplace=True)
```

```python
job.dropna(inplace=True)
```

```python

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)
job['Updated Job Title'] = job['Job Title'].apply(preprocess_text)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(job['Updated Job Title'])
def recommend_skills(job_title):
    processed_title = preprocess_text(job_title)
    title_vector = vectorizer.transform([processed_title])
    similarities = cosine_similarity(title_vector, tfidf_matrix)
    index = similarities.argmax()
    return job.iloc[index]['Key Skills']
```

```python
def final(job_title): #single input milega
    found = 0
    job_title = job_title.lower()

    if job['Updated Job Title'].str.contains(job_title).any():
            found = 1
            
    if found==1:
        recommended_skills = recommend_skills(job_title)
        result = f"{recommended_skills}"
    else:
        reuslt = "No relevant skill suggestion found"
    return result
ans = final("sales")
print(ans)
```

```python
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter


def spelling(text):
    blob = TextBlob(text)
    corrected_text = blob.correct()
    return str(corrected_text)


def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)


def find_misspelled_words(original_text, corrected_text):
    original_words = original_text.split()
    corrected_words = corrected_text.split()
    
    misspelled_words = []
    
    for original, corrected in zip(original_words, corrected_words):
        if original != corrected:
            misspelled_words.append(original)
    
    word_frequency = Counter(misspelled_words)
    return word_frequency


def number(text):
    tokens = word_tokenize(text)
    count = 0
    for token in tokens:
        try:
            int(token)
            count += 1
        except ValueError:
            continue
    return count


def score(text):
    processed_text = preprocess_text(text)    
    corrected = spelling(text)
    x = find_misspelled_words(text, corrected)
    total_misspelled = sum(x.values())
    num = number(text)
   
    final_score = -2* total_misspelled + 7 * num 
    if final_score>0:
        
        return final_score
    else:
        return "OOPS , You need to work a little more."

    

text = "Mariam Eqbal  8964079567 # eqbalmariam@gmail.com ï linkedin.com/in/mariam-eqbal § github.com/xmwriam Education Indian Institute Of Information Technology, Jabalpur August 2023 – May 2027 Bachelor of Technology in Electronics and Communication Engineering; Grade - 8.1 Madhya Pradesh, India Sushila Birla Girls’ School June 2021 – May 2023 CBSE ; Grade - 94.8% Kolkata, India Welland Gouldsmith School April 2019 – May 2021 ICSE ; Grade - 94.4% Kolkata, India Skill Summary • Languages: Python, C/C++, SQL, HTML, CSS, JavaScript, Haskell • Frameworks: Tkinter, Flask, MySQL, NumPy, Pandas, Matplotlib • Tools: VS Code, Figma, GTK, Git, Vercel, Windows, Linux, LaTeX Beamer • Soft Skills: Leadership, Event Management, Writing, Public Speaking, Time Management Projects Poetry Prompt Generator March 2024 Python, Flask, HTML, CSS, Javascript, Reddit API, Figma, Vercel Github — Website • A Web App to generate poetry prompts for writers. • Implemented a back-end that uses the Reddit API to fetch random prompts. • Wrote the frontend in HTML, CSS and Javascript, and designed in Figma. • Connected the Reddit API with the frontend using Flask and hosted the Web App on Vercel. Lifase, A Hospital Management Software October 2022 – November 2022 Python, Tkinter, MySQL Github • Tkinter was used to make an attractive and interactive GUI including a start animation. • Used MySQL to manage an extensive database handling patient, doctor, and staff records, inventory, weekly schedules, and tracked the availability of hospital beds. E-Notice boards were built for effective circulation of notices and circulars. • Generation of User and Patient ID’s was done automatically in Python on registration. An automated bill generation system was also implemented. Ducking Off - Implementing Minesweeper in Haskell June 2024 – July 2024 Haskell, GTK2 Gitlab • Two versions of the game were implemented - a CLI version and a GUI version • GTK2 was used to make the fontend of the GUI version, hence worked with legacy code • Haskell was used to handle the backend logic in both the versions. Achievements Google WE Scholar February 2024 – Present • Was selected as one of the top 200 (top 0.6%) people for WE Cohort 6 from a pool of 30,000+ candidates. Received a 100% fee scholarship along with a cash scholarship of Rs. 1,00,000. Newbie 1.0 February 2024 • Awarded Top Girl Performer in Newbie 1.0, IIIT Jabalpur’s Competitive Programming Contest out of the entire first year cohort (500+ people) of the institute, displaying my analytical skills and coding prowess. Competitive Programming November 2023 – Present • Achieved 3* on Codechef, amassing a rating of 1680. Attained global rank 196 in the contest Starters114D demonstrating my proficiency and expertise in algorithmic problem-solving. • Newbie on Codeforces with a rating of 1114. Zero Day March 2022 • Led the only School student team in a CTF (Capture The Flag) challenge organised by Department of IT and Electronics, Government of West Bengal, with 500+ participants from colleges all over West Bengal. Volunteer HackByte 2.0 January 2024 – April 2024 Volunteer IIIT Jabalpur • Spearheaded the promotion of Hackbyte 2.0, IIIT Jabalpur’s flagship Hackathon as part of the PR Team resulting in a 50% increase in participant registration from last year. A Little Contribution Welfare Society May 2022 Science Teacher Kolkata, India • Taught Science to underprivileged children as part of the ’Stars Shine for Everyone’ (SSVI) initiative under the Beamline For Schools Competition organised by CERN. Planned the syllabus and incorporated practical demonstrations and hands-on activities leading to an increased interest in STEM for 50+ students."
final_score = score(text)
print("Final Score:", final_score)
```
